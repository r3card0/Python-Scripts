## Libraries

`````python
import pandas as pd
from openpyxl.workbook import workbook
`````

* Pandas can read multiple types of excel files. This is because data it is structured in any way and you'll still be  able to interpret it
* Pandas can be read the next excel formats: xlsx,csv, txt

### 1. How to read a xlsx file

* Use ````read_excel```` function

`````python
df_excel = pd.read_excel('Exercise_Files/regions.xlsx')
df_excel
`````

* Pandas uses index as rows and headers for columns

### 2. How to read a csv file
* Use ````read_csv````

`````python
df_csv = pd.read_csv('Exercise_Files/Names.csv')
df_csv
`````

### 3. How to read a txt file
* Use ````read_csv```` function

`````python
df_txt = pd.read_csv('Exercise_Files/data.txt')
df_txt
`````

* A txt file can be read because not too much difference between CSV file and txt file except for some little formatting differences.

### 4. How to set format a txt file

* Use ````delimiter```` flag 
* This flag lets to indicate the way the columns will be separate by *tabs*

`````python
df_txt = pd.read_csv('Exercise_Files/data.txt', delimiter='\t')
df_txt
`````

## como indicar que la primera linea no es un header

* Agregando un header flag y especificar "None"

`````python
df_csv = pd.read_csv('Exercise_Files/Names.csv', header=None)
df_csv
`````

## como agregar encabezados de columnas a un archivo csv 

`````python
df_csv.columns = ['First','Last', 'Address','City','State','Area','Code']
df_csv
`````

## como convertir un archivo csv a formato xlsx

* Usar el comando: *to_excel*
* Se puede renombrar el archivo
* Dar el path del archivo

`````python
df_csv.to_excel('Exercise_Files/modified.xlsx')
`````

The rows are acting as a singular column.
